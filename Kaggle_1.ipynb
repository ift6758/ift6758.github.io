```
import nltk
nltk.download('punkt')

nltk.download('stopwords')
```

```
from nltk.corpus import stopwords 

cat_count = len(np.unique(train[1]))
frequencies = []
words = {}
categories = dict([ (cat, index) for index, cat in enumerate(np.unique(train[1]))])
for i in range(len(train[0])):
  comment_words = nltk.word_tokenize(train[0][i].lower())
  for word in comment_words:
    if word not in words:
      words[word] = len(words)
      frequencies.append([0.] * (cat_count + 2))
      frequencies[words[word]][-1] = words[word]
      
    frequencies[words[word]][-2] += 1
    frequencies[words[word]][categories[train[1][i]]] += 1
  
  
print(frequencies[0])
```
